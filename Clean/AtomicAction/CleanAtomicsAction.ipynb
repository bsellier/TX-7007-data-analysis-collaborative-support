```python
# Type de tag_name final
tag_name_categories = ["talk_to", "look_at_person", "laugh", "listen_to", "look_at_notebook_smartphone", 
                       "look_at_table", "write_on_table", "use_portal", "point_object", "look_at_board", 
                       "write_on_board", "read", "enter", "sit_down", "hand_gesture", "stand_up", "walk", 
                       "stand_board", "stand", "look_at_notebook", "look_at_slide", "talk_to_person", 
                       "talk_to_group", "point_person", "exit", "move_post_it", "take_notes"]

tag_name_type = pd.CategoricalDtype(categories=tag_name_categories)
```

```python
# Fonction de clean
def clean_atomic_actions(data_to_clean):
    data_to_clean.loc[data_to_clean['tag_name'] == 'talk to'] = 'talk_to'
    data_to_clean.loc[data_to_clean['tag_name'] == 'talk to person'] = "talk_to_person"
    data_to_clean.loc[data_to_clean['tag_name'] == 'Talk to person'] = "talk_to_person"

    data_to_clean.loc[data_to_clean['tag_name'] == 'talk to group'] = "talk_to_group"
    data_to_clean.loc[data_to_clean['tag_name'] == 'Talk to group'] = "talk_to_group"
    data_to_clean.loc[data_to_clean['tag_name'] == 'tal to group'] = "talk_to_group"

    data_to_clean.loc[data_to_clean['tag_name'] == 'look at person'] = 'look_at_person'
    data_to_clean.loc[data_to_clean['tag_name'] == 'look person'] = "look_at_person"
    data_to_clean.loc[data_to_clean['tag_name'] == 'Look person'] = "look_at_person"
    data_to_clean.loc[data_to_clean['tag_name'] == 'look at table'] = 'look_at_table'
    data_to_clean.loc[data_to_clean['tag_name'] == 'look at board'] = "look_at_board"
    data_to_clean.loc[data_to_clean['tag_name'] == 'look at slide'] = "look_at_slide"
    data_to_clean.loc[data_to_clean['tag_name'] == 'look at notebook'] = "look_at_notebook"
    data_to_clean.loc[data_to_clean['tag_name'] == 'Look at Notebook Smatphone'] = 'look_at_notebook_smartphone'

    data_to_clean.loc[data_to_clean['tag_name'] == 'write on the table'] = "write_on_table"
    data_to_clean.loc[data_to_clean['tag_name'] == 'write on table'] = "write_on_table"
    data_to_clean.loc[data_to_clean['tag_name'] == 'write on the board'] = "write_on_board"
    data_to_clean.loc[data_to_clean['tag_name'] == 'write on board'] = "write_on_board"

    data_to_clean.loc[data_to_clean['tag_name'] == 'move pot-it'] = "move_post_it"
    data_to_clean.loc[data_to_clean['tag_name'] == 'use portal'] = "use_portal"

    data_to_clean.loc[data_to_clean['tag_name'] == 'listen to'] = "listen_to"
    data_to_clean.loc[data_to_clean['tag_name'] == 'Listen to'] = "listen_to"

    data_to_clean.loc[data_to_clean['tag_name'] == 'Laugh'] = 'laugh'

    data_to_clean.loc[data_to_clean['tag_name'] == 'Reed'] = "read"
    data_to_clean.loc[data_to_clean['tag_name'] == 'reed'] = "read"

    data_to_clean.loc[data_to_clean['tag_name'] == 'Walk'] = "walk"

    data_to_clean.loc[data_to_clean['tag_name'] == 'Enter'] = "enter"
    data_to_clean.loc[data_to_clean['tag_name'] == 'Exit'] = "exit"
    data_to_clean.loc[data_to_clean['tag_name'] == 'Sit down'] = "sit_down"
    data_to_clean.loc[data_to_clean['tag_name'] == 'Stand up'] = "stand_up"
    data_to_clean.loc[data_to_clean['tag_name'] == 'Stand'] = "stand"
    data_to_clean.loc[data_to_clean['tag_name'] == 'Stand board'] = "stand_board"

    data_to_clean.loc[data_to_clean['tag_name'] == 'hand gesture'] = "hand_gesture"

    data_to_clean.loc[data_to_clean['tag_name'] == 'point at object'] = "point_object" 
    data_to_clean.loc[data_to_clean['tag_name'] == 'point person'] = "point_person"
    data_to_clean.loc[data_to_clean['tag_name'] == 'Point Person'] = "point_person"
    
    expected_size = len(tag_name_categories)
    real_size = len(all_seances.tag_name.unique().tolist())
    if real_size <= expected_size:
        print("Clean done")
    elif real_size > expected_size:
        # Si il y a plus de tag_name que définis dans le type
        print(f"Warning : Expected a maximum of {expected_size} tag names but found {real_size}")
        
    return data_to_clean

```

```python
import re
import os

# Création du dossier s'il n'existe pas

clean_folder_path = 'aa_data_clean'

if not os.path.exists(clean_folder_path):
    os.makedirs(clean_folder_path)

# Clean des data
for data in to_clean:
    data_to_clean = pd.read_csv(data)
    cleaned_data = clean_atomic_actions(data_to_clean)
    print(len(cleaned_data.tag_name.unique().tolist()))
    print(f"Length expected : {len(tag_name_categories)}")
    file_name = re.findall("([^\/]+$)", data)[0]
    print(f"Output to {clean_folder_path}/{file_name}")
    cleaned_data.to_csv(f"{clean_folder_path}/{file_name}")
```
